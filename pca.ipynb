```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.datasets import load_breast_cancer
import pandas as pd
import pennylane.numpy as np


Xt = pd.read_csv("Data/balanceado/x_train.csv")
X_test = pd.read_csv("Data/balanceado/x_test.csv")
def transform(X):
    scaler_std = StandardScaler()
    X_scaled = scaler_std.fit_transform(X)

    pca = PCA(n_components=16)
    X_pca = pca.fit_transform(X_scaled)

    quantum_scaler = MinMaxScaler(feature_range=(0, np.pi))
    X_final = quantum_scaler.fit_transform(X_pca)

    print(f"Formato original: {X.shape}")
    print(f"Formato após PCA: {X_final.shape}")
    return X_final
```

```python
X_train = transform(Xt)
X_test = transform(X_test)
```

```python
X_train = pd.DataFrame(X_train)
```

```python
X_train.to_csv("Data/balanceado/X_train.csv", index=False)
```

```python
X_test = pd.DataFrame(X_test)
X_test.to_csv("Data/balanceado/X_test.csv", index=False)
```
